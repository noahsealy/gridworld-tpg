```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_mean_std(dfs, field, length):
    arr = np.array(
            [dfs[i][field][:length] for i in range(len(dfs))])
    mean = np.mean(arr, axis=0)
    std = np.std(arr, axis=0)
    
    return mean, std
```

```python
fig11_tpg_1 = pd.read_csv('../results/fig11/tpg/0.csv')
fig11_tpg_2 = pd.read_csv('../results/fig11/tpg/1.csv')
fig11_tpg_3 = pd.read_csv('../results/fig11/tpg/2.csv')
fig11_tpg_4 = pd.read_csv('../results/fig11/tpg/3.csv')
fig11_tpg_5 = pd.read_csv('../results/fig11/tpg/4.csv')
fig11_tpgs = [fig11_tpg_1, fig11_tpg_2, fig11_tpg_3, fig11_tpg_4, fig11_tpg_5]

fig11_qtpg_1 = pd.read_csv('../results/fig11/qtpg/0.csv')
fig11_qtpg_2 = pd.read_csv('../results/fig11/qtpg/1.csv')
fig11_qtpg_3 = pd.read_csv('../results/fig11/qtpg/2.csv')
fig11_qtpg_4 = pd.read_csv('../results/fig11/qtpg/3.csv')
fig11_qtpg_5 = pd.read_csv('../results/fig11/qtpg/4.csv')
fig11_qtpgs = [fig11_qtpg_1, fig11_qtpg_2, fig11_qtpg_3, fig11_qtpg_4, fig11_qtpg_5]

fig11_lqtpg_1 = pd.read_csv('../results/fig11/lqtpg/0.csv')
fig11_lqtpg_2 = pd.read_csv('../results/fig11/lqtpg/1.csv')
fig11_lqtpg_3 = pd.read_csv('../results/fig11/lqtpg/2.csv')
fig11_lqtpg_4 = pd.read_csv('../results/fig11/lqtpg/3.csv')
fig11_lqtpg_5 = pd.read_csv('../results/fig11/lqtpg/4.csv')
fig11_lqtpgs = [fig11_lqtpg_1, fig11_lqtpg_2, fig11_lqtpg_3, fig11_lqtpg_4, fig11_lqtpg_5]
```

## All methods together, averaged over the 5 runs

```python
fig11_tpg_average_runs, fig11_tpg_average_std = get_mean_std(fig11_tpgs, 'average_score', 50)
fig11_qtpg_average_runs, fig11_qtpg_average_std = get_mean_std(fig11_qtpgs, 'average_score', 50)
fig11_lqtpg_average_runs, fig11_lqtpg_average_std = get_mean_std(fig11_lqtpgs, 'average_score', 50)

fig11_tpg_max_runs, fig11_tpg_max_runs = get_mean_std(fig11_tpgs, 'max_score', 50)
fig11_qtpg_max_runs, fig11_qtpg_max_runs = get_mean_std(fig11_qtpgs, 'max_score', 50)
fig11_lqtpg_max_runs, fig11_lqtpg_max_runs = get_mean_std(fig11_lqtpgs, 'max_score', 50)

plt.plot(fig11_tpg_average_runs, label="tpg", color='red')
plt.fill_between(range(50), fig11_tpg_average_runs-fig11_tpg_average_std, fig11_tpg_average_runs+fig11_tpg_average_std, alpha=0.2, color='red')

plt.plot(fig11_qtpg_average_runs, label="qtpg", color='blue')
plt.fill_between(range(50), fig11_qtpg_average_runs-fig11_qtpg_average_std, fig11_qtpg_average_runs+fig11_qtpg_average_std, alpha=0.2, color='blue')

plt.plot(fig11_lqtpg_average_runs, label="lqtpg", color='green')
plt.fill_between(range(50), fig11_lqtpg_average_runs-fig11_lqtpg_average_std, fig11_lqtpg_average_runs+fig11_lqtpg_average_std, alpha=0.2, color='green')

plt.title("Run Comparisons")
plt.xlabel("Generation")
plt.xlim(0,50)
plt.ylabel("Fitness")
plt.ylim(-10,130)
plt.yticks(np.arange(0, 131, 25))
plt.legend()

plt.show()
```
