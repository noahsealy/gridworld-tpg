```python
import random
records = []
for i in range(100):
    if (i % 5 == 0):
        agent = 5
    else:
        agent = random.randrange(0, 100000000)
    records.append({'agent':agent, 'learner':random.randrange(0, 100000000)})
for record in records:
    print(record)
```
